<b>Executing Flair NER models</b>

This notebook documents how to generate NER predictions aligned with the ArchOnto ontology from datasets available at: [https://doi.org/10.6084/m9.figshare.28970633](https://doi.org/10.6084/m9.figshare.28970633).

<b>Import the required libraries</b>

```python
from flair.data import Corpus
from flair.datasets import ColumnCorpus
from flair.models import SequenceTagger
import time
```

<b>Load ground-truth files</b>

```python
# Dataset options https://doi.org/10.6084/m9.figshare.28970633
src_path="ner/datasets/ner-domain-gen/"

columns = {0: 'text', 1: 'ner'}
corpus: Corpus = ColumnCorpus(src_path, columns, test_file='test.tsv')
```

<b>Load Flair model</b>

```python
# Flair model options in https://doi.org/10.6084/m9.figshare.28970633
flair_model="ner/models/flairel+word2vec"
model = SequenceTagger.load(flair_model)

```

<b>Execute Flair model and save prediction file</b>

```python
dst_path="path/to/your/output/test.tsv"

start_time = time.time()
result = model.evaluate(corpus.test, gold_label_type="ner",dst_path)
end_time = time.time()
```

<b>Print execution inference time</b>

```python
print(f"Total runtime of the program is {end_time - start_time:.0f} seconds")
```
